#### Outliers
There is a possibility of outliers in this dataset given that there might be instances where the Uber was ordered in the outskirts or at a location thats usually abandon. We do not want such outlier to interfere.
Because Latitude and Longitude are geographic points, we need to use DBSCAN to find outliers. Due to hardware issues, here is an shorter version.

```python
# Creating Radians
df['Lat_rad'] = np.radians(df['Lat'])
df['Lon_rad'] = np.radians(df['Lon'])

tree = BallTree(df[['Lat_rad', 'Lon_rad']].values, metric='haversine')

eps = 20  
min_samples = 1000

# DBSCAN
labels = tree.query_radius(df[['Lat_rad', 'Lon_rad']].values, r=eps, count_only=True)
core_points = np.where(labels >= min_samples)[0]
outlier_points = np.where(labels < min_samples)[0]


df['Cluster_Label'] = -1  # Initializing with noise label
df['Cluster_Label'].iloc[core_points] = 1  # Assign core point label

df_no_outliers = df[df['Cluster_Label'] == 1]

df_no_outliers.drop(columns=['Lat_rad', 'Lon_rad', 'Cluster_Label'], inplace=True)

print(df_no_outliers)
```

There were no Outliers because the number of rows is unchanged

## First Model: K nearest neighbor Classifier

Finding the optimal number of clusters and the locations of them using *Elbow Method*

```python
# Define the range of clusters to try
k_range = list(range(1, 11))

# Initialize an empty list to store the scores
scores = []

# Run KMeans for each number of clusters
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    score = kmeans.inertia_  # sum of squared distances to the closest centroid for all observations in the cluster
    scores.append(score)

# Plot the scores
plt.plot(k_range, scores, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method')
plt.show()
```

We can see that the ideal number of centres as per Elbow Method is 10

```python
# Choose the number of clusters as the one that gives the "elbow" of the plot
n_clusters = scores.index(min(scores)) + 1
```

```python
# Run KMeans with the chosen number of clusters
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
kmeans.fit(X)

# Get the coordinates of the cluster centers
centers_kmeans = kmeans.cluster_centers_
```

```python
# Create a map centered around New York
map_nyc = folium.Map(location=[40.7128, -74.0060], zoom_start=10)

# Add the cluster centers to the map
for center in centers_kmeans:
    folium.CircleMarker(
        location=center,
        radius=5,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_nyc)

# Display the map
map_nyc
```

## Second Model: GMM

```python
# Lists to hold the BIC (Bayesian Information Criterion) values
bic = []

# Looping over multiple values of k to find optimal
for k in range(2, 10):
    # Define a GMM model
    gmm = GaussianMixture(n_components=k, random_state=42)
    
    # Fit the data to the model
    gmm.fit(X)
    
    # Append the BIC to the list
    bic.append(gmm.bic(X))

# Plot the BIC values
plt.plot(range(2, 10), bic, marker='o')
plt.title('GMM - BIC for different number of clusters')
plt.xlabel('Number of clusters')
plt.ylabel('BIC')
plt.show()

```

```python
# Choose the number of components at the lowest BIC
n_components = bic.index(min(bic)) + 2

# Build a GMM with the optimal number of components
gmm = GaussianMixture(n_components=n_components, random_state=42)

# Fit the data
gmm.fit(X)

# Predict the cluster for each data point
y_gmm = gmm.predict(X)

# Print the cluster centers
print("Optimal number of components: ", n_components)
print("Means of the components: ")
print(gmm.means_)
```

## Model Evaluation

When comparing clustering models like K-Means and GMM, we can use several metrics. Here are a few:

Inertia: This is the sum of squared distances of samples to their closest cluster center. A lower inertia value is better. However, inertia makes the assumption that clusters are convex and isotropic, which is not always the case. It responds poorly to elongated clusters, or manifolds with irregular shapes.

Silhouette Score: This measures how close each sample in one cluster is to the samples in the neighboring clusters. Its value ranges from -1 to 1, where a higher value indicates that the sample is closer to its own cluster than to other clusters. This can be used with any distance metric, such as Euclidean distance or Manhattan distance.

Calinski-Harabasz Index: Also known as the Variance Ratio Criterion, this index is the ratio of the sum of between-clusters dispersion and of inter-cluster dispersion for all clusters (where dispersion is defined as the sum of distances squared). Higher values indicate better clustering.

Davies-Bouldin Index: This index signifies the average 'similarity' between clusters, where the similarity is a measure that compares the distance between clusters with the size of the clusters themselves. Zero is the lowest possible score. Values closer to zero indicate a better partition.
